## similarity fitted optic flow field + estimated RF

```python
def FE_similarity(F, E):
    return ((np.sum(F*E, axis=1)/
            (np.linalg.norm(np.clip(E, 0.0000001, 1.), axis=1) *
             np.linalg.norm(F, axis=1)))
            .mean())
```
